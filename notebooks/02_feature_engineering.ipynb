### PAYMENT CAPACITY FEATURES

```python
risk["PAY_RATIO_LATEST"] = np.where(
    risk["BILL_AMT1"] > 0,
    risk["PAY_AMT1"] / risk["BILL_AMT1"],
    0
    )
```

```python
risk[["PAY_RATIO_LATEST", "default.payment.next.month"]].corr()
```

```python
pay_cols = ["PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]
bill_cols = ["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]

ratios = risk[pay_cols].values / risk[bill_cols].values
ratios = np.nan_to_num(ratios, nan=0.0, posinf=0.0, neginf=0.0)
```

```python
risk["PAY_RATIO_MEAN"] = ratios.mean(axis=1)
```

```python
risk[["PAY_RATIO_MEAN", "default.payment.next.month"]].corr()
```

```python
ratios = risk[pay_cols].values / risk[bill_cols].values
ratios[risk[bill_cols].values <= 0] = np.nan
risk["PAY_RATIO_MEAN"] = np.nanmean(ratios, axis=1)
```

```python
risk[["PAY_RATIO_MEAN", "default.payment.next.month"]].corr()
```

### PROP MONTHS LOW PAYMENT

```python
low_payment = ratios < 0.5
low_payment = low_payment.astype(int)
valid_months = ~np.isnan(ratios)
```

```python
risk["PROP_MONTHS_LOW_PAYMENT"] = (
    low_payment.sum(axis=1) /
    valid_months.sum(axis=1)
)
```

```python
risk[["PROP_MONTHS_LOW_PAYMENT", "default.payment.next.month"]].corr()
```

### PROP_MONTHS_ZERO_PAYMENT

```python
zero_payment = (risk[pay_cols].values == 0) & (risk[bill_cols].values > 0)
risk["PROP_MONTHS_ZERO_PAYMENT"] = (zero_payment.sum(axis=1)) / (valid_months.sum(axis=1))
```

```python
risk[["PROP_MONTHS_ZERO_PAYMENT", "default.payment.next.month"]].corr()
```

### PROPORTION OF MONTHS WITH OVERPAYMENT

```python
over_payment = ratios > 1

risk["PROP_MONTHS_OVERPAYMENT"] = (
    over_payment.sum(axis=1) /
    valid_months.sum(axis=1)
)
```

```python
risk[["PROP_MONTHS_OVERPAYMENT", "default.payment.next.month"]].corr()
```

### PAY_DISCIPLINE_SCORE

```python
risk["PAY_DISCIPLINE_SCORE"] = (
    1.0 * risk["PROP_MONTHS_OVERPAYMENT"]
    - 1.0 * risk["PROP_MONTHS_LOW_PAYMENT"]
    - 1.5 * risk["PROP_MONTHS_ZERO_PAYMENT"]
)
```

```python
risk[["PAY_DISCIPLINE_SCORE", "default.payment.next.month"]].corr()
```
