```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
risk = pd.read_csv("../data/UCI_Credit_Card.csv")
risk.head()
```

### PAY LATEST

```python
risk["PAY_LATEST"] = risk["PAY_0"]
```

```python
risk[["PAY_LATEST", "default.payment.next.month"]].head()
```

```python
risk[["PAY_LATEST", "default.payment.next.month"]].corr()
```

### PAY MEAN

```python
risk["PAY_MEAN"] = risk[["PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]].mean(axis=1)
```

```python
risk[["PAY_MEAN", "default.payment.next.month"]].head()
```

```python
risk[["PAY_MEAN", "default.payment.next.month"]].corr()
```

### PAY MAX DELAY

```python
risk["PAY_MAX_DELAY"] = risk[["PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]].max(axis=1)
risk[["PAY_MAX_DELAY", "default.payment.next.month"]].head()
```

```python
risk[["PAY_MAX_DELAY", "default.payment.next.month"]].corr()
```
